````python
# Query to calculate the statistics of housing dataset ,
query = """
SELECT
    ZIP AS ZIP_Code,
    COUNT(*) AS Total_Projects,
    SUM(SITE_UNITS) AS Total_Site_Units,
    AVG(SITE_UNITS) AS Avg_Site_Units,
    SUM(PROJECT_TOTAL_UNITS) AS Total_Project_Units,
    AVG(PROJECT_TOTAL_UNITS) AS Avg_Project_Units,
    SUM(SH_UNITS_PER_SITE) AS Total_SH_Units,
    AVG(SH_UNITS_PER_SITE) AS Avg_SH_Units,
    SUM(TAX_EXEMPT_CONDUIT_BOND) AS Total_Tax_Exempt_Bonds,
    AVG(TAX_EXEMPT_CONDUIT_BOND) AS Avg_Tax_Exempt_Bonds,
    SUM(JOBS) AS Total_Jobs,
    AVG(JOBS) AS Avg_Jobs
FROM
    housing_prices
GROUP BY
    ZIP
"""

result_df = pd.read_sql_query(query, conn)
````

````python
from scipy.stats import pearsonr
````

````python
# Define predictor variables (related to affordable housing statistics)
predictor_vars = [
    'Total_Site_Units', 'Avg_Site_Units', 'Total_Project_Units',
    'Avg_Project_Units', 'Total_SH_Units', 'Avg_SH_Units',
    'Total_Tax_Exempt_Bonds', 'Avg_Tax_Exempt_Bonds',
    'Total_Jobs', 'Avg_Jobs'
]

# Define predicted variables related to health inspection scores
predicted_vars = ['Avg_Score', 'Max_Score']

# Perform Pearson correlation test for each predictor with health-related variables
results = []
for predictor in predictor_vars:
    for predicted in predicted_vars:
        # Drop NaN values from both predictor and predicted columns at the same time
        data = result_df[[predictor, predicted]].dropna()

        # Check if there are enough data points to compute correlation
        if len(data) > 1:  # Pearson requires at least 2 data points
            corr, p_value = pearsonr(data[predictor], data[predicted])
            # Append the results
            results.append({
                'Predictor': predictor,
                'Predicted': predicted,
                'Pearson Correlation': corr,
                'P-Value': p_value
            })

# Convert results into a DataFrame for easy viewing
results_df = pd.DataFrame(results)

# Set the significance level
alpha = 0.05

# Filter out the significant correlations
significant_results = results_df[results_df['P-Value'] < alpha]

# Display the significant correlations
print("Significant Correlations with Health Inspection Scores:")
print(significant_results[['Predictor', 'Predicted', 'Pearson Correlation', 'P-Value']])

# Interpretation: Identify significant correlations
for index, row in significant_results.iterrows():
    print(f"Significant correlation found between {row['Predictor']} and {row['Predicted']}:")
    print(f"Pearson Correlation: {row['Pearson Correlation']:.3f}, P-Value: {row['P-Value']:.5f}")
````

````python
# Calculate means for Northern and Southern states
north_soil_moisture = df_filtered[df_filtered['Region'] == 'North']['Avg_smlvl_at15cm']
south_soil_moisture = df_filtered[df_filtered['Region'] == 'South']['Avg_smlvl_at15cm']

# If data is not normally distributed, we use Mann-Whitney U test
stat, p = mannwhitneyu(north_soil_moisture, south_soil_moisture)
print(f"Mann-Whitney U Test Results:\nStatistic = {stat}\nP-Value = {p}")
````
